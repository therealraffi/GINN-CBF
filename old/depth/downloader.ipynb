```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
name = "brown_cogsci_1"
website_url = f"https://sun3d.cs.princeton.edu/data/{name}/{name}/"
output_dir = f"../datasets/SUN3d/{name}/"

dir_names = ["", "image", "depth", "extrinsics", "annotation"]
num_files = 500
```

```python
os.makedirs(output_path, exist_ok=True)

for dirname in dir_names:
    directory_url = website_url + dirname + "/"
    output_path = output_dir + dirname

    response = requests.get(directory_url)
    soup = BeautifulSoup(response.text, "html.parser")

    files = [link.get('href') for link in soup.find_all('a') if link.get('href').endswith(('jpg', 'jpeg', 'png', 'txt', 'json'))]
    files = files[:num_files]

    for file in files:
        file_url = directory_url + file
        print(f"Downloading {file_url}")
        !wget -P {output_path} {file_url}
```
